```python
import polars as pl
import matplotlib.pyplot as plt

# в‘  жЁ№зЁ®гҒ”гҒЁгҒ®гӮ«гӮҰгғігғҲ
df_count = data_pl.group_by("жЁ№зЁ®").len()

# в‘Ў еүІеҗҲпјҲгғ‘гғјгӮ»гғігғҶгғјгӮёпјүгҒ«еӨүжҸӣ
df_ratio = df_count.with_columns(
    (pl.col("len") / pl.col("len").sum() * 100).alias("percentage")
)

# numpyгҒёеӨүжҸӣ
labels = df_ratio["жЁ№зЁ®"].to_list()
sizes = df_ratio["percentage"].to_list()

# в‘ў еҶҶгӮ°гғ©гғ•
plt.figure()
plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("жЁ№зЁ®гҒ”гҒЁгҒ®еүІеҗҲ")
plt.show()
```

```python
species_list = data_pl["жЁ№зЁ®"].unique().to_list()
```

```python
species_list
```

```python
cols = [c for c in data_pl.columns if c not in ["жЁ№зЁ®", "id"]]

data_pl.filter(pl.col("жЁ№зЁ®") == "гӮӨгғҒгғ§гӮҰ")\
    .with_columns(
        pl.sum_horizontal(
            [pl.col(c).is_not_null() for c in cols]
        ).alias("cnt")
    )\
    .select([
        pl.col("cnt").min().alias("min_cnt"),
        pl.col("cnt").max().alias("max_cnt")
    ])
```

```python
for timber in species_list:
    result = data_pl.filter(pl.col("жЁ№зЁ®") == timber)\
    .with_columns(
        pl.sum_horizontal(
            [pl.col(c).is_not_null() for c in cols]
        ).alias("cnt")
    )\
    .select([
        pl.col("cnt").min().alias("min_cnt"),
        pl.col("cnt").max().alias("max_cnt")
    ])
    print(timber)
    print(result)
```
